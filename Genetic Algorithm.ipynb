```python
import random 

#The board size
print("The size of the board is 255 X 255")
print(" ")

#Population
population = 6
point_num =int(input("Please specify the number of points you would like to have in the polygon : " ))
print(" ")

polygons = []

#All points generation for the polygon
for i in range(point_num):
    x_axis = []
    y_axis = []
    # Generating points as of binary 8-bit X and Y 
    for i in range(8):
        x = random.randint(0,1)   
        x_axis.append(x)   
    
    for i in range(8):
        y = random.randint(0,1)
        y_axis.append(y)
    
    polygons.append(x_axis)
    polygons.append(y_axis) 
    
   # print("x" , x_axis)
   # print("y" , y_axis)
    
print(" ")    
print(polygons)

#Binary to decimal conversion for plotting on graph
dcml_value = 0
decimal_points = []
total = 7

for j in range(0 , point_num*2):
    dcml_value = 0
    for i in range( 0 , 8):
        digit = polygons[j][i]
        
        if (digit == 1):
            dcml_value = dcml_value + pow(2 , total-i)
            
    decimal_points.append(dcml_value)
    
print(" ")
      
print("Decimal values : " , decimal_points)      

# importing the required module
import matplotlib.pyplot as plt

#Plotting points of the polygon in the graph
x =[]
y =[]

for i in range(0 , point_num*2):
    
    if(i%2==0):
        # x axis values
        x.append(decimal_points[i])
    elif(i%2!=0):    
        # corresponding y axis values
        y.append(decimal_points[i])
 
x.append(decimal_points[0])
y.append(decimal_points[1])

#plotting the points
plt.plot(x, y)
 
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# giving a title to my graph
plt.title('POLYGON')
 
# function to show the plot
plt.show()

import math

# FITNESS FUNCTION
#----------------
def findequation(x1,y1, x2,y2):
    x1 = x1[0]
    x2 = x1[1]
    y1 = y1[0]
    y2 = y1[1]
    m1 = (x2-x1)/(y2-y1)
    yinter1 = y1 - (m1*x1)
    # y = mx + c
    # y = m1(x) - yinter1

    a1 = x2[0]
    b1 = x2[1]
    a2 = y2[0]
    b2 = y2[1]
    m2 = (a2-a1)/(b2-b1)
    yinter2 = b1 - (m2*a1)
  #equation of this line -> y = m2(x) - yinter2
  
  # m1(x) - yinter1 = m2(x) - yinter2
  #calculate the calue for x in above equation. this gives us the intersection x coordinate
  #
  #find the value of y coordinate but putting the above calculated x value in one of the equations above

size = len(decimal_points)
tempx = []
tempy = []
tempx.append(decimal_points[0])
tempx.append(decimal_points[2])
tempy.append(decimal_points[1])
tempy.append(decimal_points[3])

for i in range(0,size-2,2):
    tempx[0] = decimal_points[i]
    tempx[1] = decimal_points[i+2]
    tempy[0] = decimal_points[i+1]
    tempy[1] = decimal_points[i+3]
  
    print(tempx, end = '')
    print(tempy, end = '')
    plt.plot(tempx, tempy)
  
tempx[0] = tempx[1]
tempx[1] = decimal_points[0]
tempy[0] = tempy[1]
tempy[1] = decimal_points[1]
print(tempx, end = '')
print(tempy, end = '')
plt.plot(tempx, tempy)

plt.show()
intersects = []

for i in range(population):  
    num = random.randint( 0 , population/2)
    intersects.append( num )

# Roulette Selection
total = 0
selection = []
for i in range (0 , population):
    total += intersects[i] 
    
print(total)     

for i in range( 0 , population):
    intersects[i] = intersects[i] / total
    
print(intersects)  

for j in range(0 , population):   
    rand_num = random.randint(0, population-1)
    selection.append(intersects[rand_num])
    
print("Roulette selection : " ,selection)    

#Crossover 

#Generating random points of crossing over 

for i in range(8):
    
    k = random.randint(0, 7) 

    temp = polygons[i][k]
    polygons[i][k] = polygons[i+2][k]
    polygons[i+2][k] = temp

print(" ")    
print("After Crossover : " ,polygons)       

#Mutation
for i in range(8):
    m = random.randint(0 , 7)

    if(polygons[i][m] == 0):
        polygons[i][m] = 1
        
    elif(polygons[i][m] == 1):
        polygons[i][m] = 0
        
        
print(" ")    
print("After Mutation : " ,polygons)         
```
